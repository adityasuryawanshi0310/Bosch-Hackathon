```python
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov8n.pt")  # Use "yolov8s.pt" or "yolov8m.pt" for better accuracy

# Train the model
model.train(
    data="/content/unzipped_folder/bsgw/crosswalk/data.yaml",  # Path to your data.yaml
    epochs=100,
    imgsz=640,
    batch=32,  # Adjust based on GPU memory (T4 can handle 32-64)
    device=0,  # Use the free T4 GPU
    name="crosswalk_detection",
    optimizer="Adam",  # Faster convergence
    lr0=0.001,  # Learning rate
    plots=True  # Generate training plots
)
```

```python

model = YOLO("/content/runs/detect/crosswalk_detection/weights/best.pt")

model.save("best_crosswalk.pt")
```
